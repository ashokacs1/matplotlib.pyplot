## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. The data is represented as points on a Cartesian coordinate system, where the position of each point on the horizontal (x) and vertical (y) axes indicates values for each of the two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("this are x coordinate values")
plt.ylabel("this are y co oridinate values")
plt.title("x axis  vs y axis")
plt.grid()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np

## ==>>
The subplot() function in Matplotlib is used to create multiple plots in a single figure. This is particularly useful when you want to compare different datasets or visualizations side by side, within the same figure.

Why is the subplot() function used?
Multiple Plots in One Figure: It allows you to create a grid of plots within one figure, making it easier to analyze and compare multiple plots together.
Efficient Space Management: Subplots help in managing space effectively by arranging multiple plots in rows and columns within a single figure.
Custom Layouts: You can create custom layouts for your plots, such as a 2x2 grid, a 3x1 grid, etc., depending on your requirements.
Drawing Four Line Plots using the subplot() Function:

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating a figure with subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.title('Line 1')
plt.xlabel('x')
plt.ylabel('y')

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.title('Line 2')
plt.xlabel('x')
plt.ylabel('y')

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b')
plt.title('Line 3')
plt.xlabel('x')
plt.ylabel('y')

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'k')
plt.title('Line 4')
plt.xlabel('x')
plt.ylabel('y')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

Explanation:

plt.subplot(2, 2, 1): Creates a 2x2 grid of subplots and selects the first subplot.

plt.plot(x, y1, 'r'): Plots the first line in red ('r') on the first subplot.

plt.subplot(2, 2, 2): Selects the second subplot in the grid.

plt.plot(x, y2, 'g'): Plots the second line in green ('g') on the second subplot.

plt.subplot(2, 2, 3): Selects the third subplot in the grid.

plt.plot(x, y3, 'b'): Plots the third line in blue ('b') on the third subplot.

plt.subplot(2, 2, 4): Selects the fourth subplot in the grid.

plt.plot(x, y4, 'k'): Plots the fourth line in black ('k') on the fourth subplot.

plt.tight_layout(): Adjusts the spacing between subplots to prevent overlapping.

This code will produce a figure with four line plots arranged in a 2x2 grid.








## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot (or bar chart) is a type of data visualization that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots can be plotted either vertically (with bars standing upright) or horizontally (with bars lying flat).

Why is a Bar Plot Used?
Comparison Between Categories: Bar plots are ideal for comparing the sizes or quantities of different categories or groups. Each bar represents a different category, making it easy to see how they stack up against one another.

Visualizing Discrete Data: Bar plots are used when the data is discrete, meaning it represents distinct and separate categories or groups rather than continuous data.

Highlighting Differences: They effectively highlight differences in the magnitude of the categories, making it easy to spot which category is the largest or smallest.

Simple and Easy to Read: Bar plots are simple to create and easy to interpret, making them one of the most commonly used types of plots in data analysis and reporting.

Example Use Cases:
Sales Data: Comparing the sales figures of different products.
Population Data: Showing the population of different regions or countries.
Survey Results: Visualizing the number of responses for different survey options.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company names")
plt.ylabel("companies profit")
plt.title("the vertical bar chart")
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel("company names")
plt.ylabel("companies profit")
plt.title("the horizantal bar chart")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary:

 the minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

 It is used to show the spread and skewness of data, as well as to identify potential outliers.

Why is a Box Plot Used?
Visualizing Distribution:

 Box plots provide a visual summary of the data distribution, including its central tendency, variability, and symmetry or skewness.

Identifying Outliers:

Box plots help in detecting outliers, which are data points that fall outside the typical range of the data.

Outliers are shown as individual points outside the "whiskers" of the plot.

Comparing Multiple Datasets:

 Box plots are useful for comparing the distributions of multiple datasets side by side, making it easier to see differences in their medians, spreads, and the presence of outliers.

Summarizing Data:

They provide a quick summary of the data with minimal visual clutter, making them ideal for exploratory data analysis.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2],labels=['Box1','Box2'])
plt.title("the boxplot")
plt.show()
```
